## Classification of random data

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.datasets import make_circles, make_blobs, load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Generate random data 
X, Y = make_blobs(n_features=2, centers=3)

plt.scatter(X[Y==0, 0], X[Y==0, 1], marker='o', s=50, edgecolor='k')
plt.scatter(X[Y==1, 0], X[Y==1, 1], marker='o', s=50, edgecolor='k')
plt.scatter(X[Y==2, 0], X[Y==2, 1], marker='o', s=50, edgecolor='k')
plt.show()
```

```python
neigh = KNeighborsClassifier(n_neighbors=15)
neigh.fit(X, Y)
```

```python
def draw_border(X, Y):
    x_min, x_max = X[:, 0].min() - 0.3, X[:, 0].max() + 0.3
    y_min, y_max = X[:, 1].min() - 0.3, X[:, 1].max() + 0.3
    h = 0.1

    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    xx_ = np.reshape(xx, (xx.shape[0]*xx.shape[1],1))
    yy_ = np.reshape(yy, (yy.shape[0]*yy.shape[1],1))

   
    X1 = (np.c_[xx_.flatten(), yy_.flatten()])
    global pred
    pred = []

    pred = neigh.predict(X1)

    pred = np.asarray(pred)
    pred = np.reshape(pred, (xx.shape[0], xx.shape[1]))

    fig = plt.figure()
    plt.contourf(xx, yy, pred, cmap='jet', alpha=0.8)

    plt.scatter(X[Y==0, 0],X[Y==0, 1], color='yellow', edgecolor='k', label='Category: 0')
    plt.scatter(X[Y==1, 0],X[Y==1, 1], color='r', edgecolor='k', label='Category: 1')
    plt.scatter(X[Y==2, 0],X[Y==2, 1], color='lime', edgecolor='k', label='Category: 2')
    plt.legend()
    plt.show()

draw_border(X, Y)
```

```python
pred = neigh.predict(X)
print('accuracy: ', accuracy_score(Y, pred)*100, '%')
```

## Iris dataset classification


```python
iris = load_iris()
X = iris.data[:, :2]
Y = iris.target
```

```python
plt.scatter(X[Y==0, 0], X[Y==0, 1], marker='o', s=50, edgecolor='k')
plt.scatter(X[Y==1, 0], X[Y==1, 1], marker='o', s=50, edgecolor='k')
plt.scatter(X[Y==2, 0], X[Y==2, 1], marker='o', s=50, edgecolor='k')
plt.show()
```

```python
neigh = KNeighborsClassifier(n_neighbors=15)
neigh.fit(X, Y)
```

```python
def draw_border(X, Y):
    x_min, x_max = X[:, 0].min() - 0.3, X[:, 0].max() + 0.3
    y_min, y_max = X[:, 1].min() - 0.3, X[:, 1].max() + 0.3
    h = 0.1

    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    xx_ = np.reshape(xx, (xx.shape[0]*xx.shape[1],1))
    yy_ = np.reshape(yy, (yy.shape[0]*yy.shape[1],1))

   
    X1 = (np.c_[xx_.flatten(), yy_.flatten()])
    global pred
    pred = []

    pred = neigh.predict(X1)

    pred = np.asarray(pred)
    pred = np.reshape(pred, (xx.shape[0], xx.shape[1]))

    fig = plt.figure()
    plt.contourf(xx, yy, pred, cmap='hot', alpha=0.8)

    plt.scatter(X[Y==0, 0],X[Y==0, 1], color='yellow', edgecolor='k', label='Category: 0')
    plt.scatter(X[Y==1, 0],X[Y==1, 1], color='skyblue', edgecolor='k', label='Category: 1')
    plt.scatter(X[Y==2, 0],X[Y==2, 1], color='lime', edgecolor='k', label='Category: 2')
    plt.legend()
    plt.show()

draw_border(X, Y)
```
